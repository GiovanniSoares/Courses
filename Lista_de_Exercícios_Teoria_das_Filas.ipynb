1. Para realizar a validação operacional de um modelo de simulação, deseja-se determinar se o modelo tem o mesmo desempenho do nível de atendimento real. A tabela abaixo ilustra 12 dados reais do nível de atendimento e 12 replicações do modelo de simulação para uma taxa de chegada de chamada de 100 clientes por hora.
image.png

a) 95% - alpha=0.05 
b) 99% - alpha=0.01

```
real=[13, 12, 16, 14, 15, 10, 13, 11]
simulado=[14, 17, 14, 16, 15, 17, 12, 19]
alpha=0.05
```

```
                      
df = (2*len(real))-2                                       
t = sts.t.ppf(1 - alpha/2,df)                             
LI = np.mean(simulado) - np.mean(real) - t * (np.sqrt(((statistics.pstdev(simulado))**2+(statistics.pstdev(real))**2)/len(real)))
LS = np.mean(simulado) - np.mean(real) + t * (np.sqrt(((statistics.pstdev(simulado))**2+(statistics.pstdev(real))**2)/len(real)))
```

```
(LI,LS)
```

COMO O INTERVALO CONTEM O VALOR ZERO PODEMOS AFIRMAR  COM 99% DE CONFIANÇA QUE O MODELO É EQUIVALE AO SISTEMA REAL

4. Em um estudo de simulação, deseja-se analisar o comportamento das pessoas ao atravessarem uma determinada via. Os valores abaixo representam os valores obtidos, em minutos, para os tempos que as pessoas levaram na travessia (tempo de espera + caminhada pela via).

0,95 0,85 0,92 0,95 0,93 0,86 1,00 6,10 0,92 0,85 0,81
0,78 0,93 0,93 1,05 0,93 0,10 1,06 1,06 0,96 0,81 0,96

Construa os intervalos de confiança para a média da população, considerando confianças estatísticas de 99% e 95%.

a) 99% - alpha=0.01
b) 95% - alpha = 0.05

```
tempo=[0.95, 0.85, 0.92, 0.95, 0.93, 0.86, 1.00, 6.10, 0.92, 0.85, 0.81,
0.78, 0.93, 0.93, 1.05, 0.93, 0.10, 1.06, 1.06, 0.96, 0.81, 0.96]
alpha=0.01
```

```
s=statistics.stdev(tempo)

n=(len(tempo))

df = n-1                                       
t = sts.t.ppf(1 - alpha/2,df)                             
LI = np.mean(tempo) - t *(s/(np.sqrt(n)))
LS= np.mean(tempo) + t *(s/(np.sqrt(n)))
```

```
(LI,LS)
```
